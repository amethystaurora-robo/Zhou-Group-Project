```python
import pandas as pd
import seaborn as sns
import numpy as np
import re
from scipy.stats import chi2_contingency
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, roc_curve, auc
```

```python
LCMS_df = pd.read_table('LCMS Target Screening detections raw data 2023-08-02.tsv')
GCMS_df = pd.read_table('GCMS Target Screening detections raw data 2023-08-02.tsv')
all_genus_df = pd.read_csv('all_genus/Aquatic-Export-Table 1.csv')
concentration_df = pd.read_csv('quantified_concentration/Aquatic-Export-Table 1.csv')
population_df = pd.read_csv('more_fish/population.csv')
anglia_coords = pd.read_csv('anglia_coords.csv')
```

```python
LCMS_GCMS_df = pd.concat((LCMS_df, GCMS_df), axis = 0)

LCMS_GCMS_df.head()

#make pesticide df
unique_uses = LCMS_GCMS_df["USE"].tolist()

simplified_usage_list = []
patternP='DEET|Pesticide|Intermediate; Pesticides|Insecticide|2,4-D preparation|Biocide|Plant'

for chemical in unique_uses:
    if re.search(patternP,chemical):
        simplified_usage_list.append('Pesticide or Insecticide')
    else:
        simplified_usage_list.append('Other')


LCMS_GCMS_df['Usages'] = simplified_usage_list

#subset for just pesticide data
pesticides_df = LCMS_GCMS_df.loc[LCMS_GCMS_df['Usages']=='Pesticide or Insecticide']
```

```python
#add column with pesticide subset
conazoles = 'azole|chlorophenyl[)]|phenoxy|diflu|butanil|chloraz|SN|Triad|izole|Imaza|Flut|fluorop'
cyanide = 'cyanide|Hydroxy|Acryl|Cyanide|cyan|Benzene|Propen'
ddt = 'DD|Perthane|benzene'
dibenzo = 'enzofuran|2,3'
neocotinoids = 'thia|Thia|clop|cloth|pai|iazi|pyram|inot|mip|imidazo|guanidine'
perchlorates = 'perch|Perch'
strobins = 'strob|mis|ativ|pera|brio|CST|uilt|tego|sox|methoxycarbamic'

pesticide_list = []
compounds_list = all_genus_df['Chemical Name'].tolist()
pests_list = pesticides_df['Compound_Name'].tolist()
pests_empty_list = []

def sort_pests(df_list,empty_list):
    for chemical in df_list:
        if re.search(conazoles,chemical):
            empty_list.append('Azole')
        elif re.search(cyanide,chemical):
            empty_list.append('Cyanide')
        elif re.search(ddt,chemical):
            empty_list.append('DDT')
        elif re.search(dibenzo,chemical):
            empty_list.append('Dibenzofuran')
        elif re.search(neocotinoids,chemical):
            empty_list.append('Neonicotinoid')
        elif re.search(perchlorates,chemical):
            empty_list.append('Perchlorate')
        elif re.search(strobins,chemical):
            empty_list.append('Strobilurin')
        else:
            empty_list.append('Other')

    return empty_list

sort_pests(compounds_list,pesticide_list)
all_genus_df['Pesticide Categories'] = pesticide_list
#concentration_df['Pesticide Categories'] = pesticide_list

sort_pests(pests_list,pests_empty_list)
pesticides_df['Pesticide Categories'] = pests_empty_list
pesticides_df.head()
```
